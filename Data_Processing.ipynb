```python
playerid_df = pd.read_sql_query("SELECT * FROM Player", cnx)
playerid_df.head()
```

```python
mergedplayer_df = pd.merge(df_unique_players, playerid_df, left_on='Name', right_on='player_name', how='inner')  # Use 'left', 'right', or 'outer' for different types of joins
mergedplayer_df.head()
```

```python
def add_player_stats(row):
  new_row = row.copy()
  for col in row.index:
    if 'player' in col:
      player_id = row[col]
      player_stats = mergedplayer_df[mergedplayer_df['player_id'] == player_id].drop(columns=['id_y']).iloc[0]
      for stat in player_stats.index:
        new_col_name = f"{col.split('_code')[0]}_{stat}"
        new_row[new_col_name] = player_stats[stat]
  return new_row
result = match_df_16[:5]
match_df_16.head()
enhanced_matches_df = result.apply(add_player_stats, axis=1)
enhanced_matches_df.head()
#enhanced_matches_df = matches_df.apply(add_player_stats, axis=1)
```
